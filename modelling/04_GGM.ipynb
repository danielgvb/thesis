# GGM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
from sklearn.metrics import mean_squared_error, r2_score
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('../Data/silver/covid_data_weekly.csv')
df.head()
```

```python
df_col = df[df.country=='Colombia'].reset_index(drop=True)


df_col['date'] = pd.to_datetime(df_col['date'])
df_col.set_index('date', inplace = True)
covid_series = df_col.cases
covid_series = covid_series[covid_series!=0]
covid_series
```

```python
# Define the GGM model function
def ggm_model(t, K, pc, qc, ps, qs):
    # Calculate the model expression
    part1 = np.sqrt((1 - np.exp(-(pc + qc) * t)) / (1 + (qc / pc) * np.exp(-(pc + qc) * t)))
    part2 = (1 - np.exp(-(ps + qs) * t)) / (1 + (qs / ps) * np.exp(-(ps + qs) * t))
    return K * part1 * part2

# Function to fit the GGM model to a time series
def fit_ggm_model(time_series):
    # Extract time and values from the time series
    t = time_series.index.values
    z = time_series.values

    # Initial guesses for the parameters
    initial_guess = [np.max(z), 0.01, 0.1, 0.01, 0.1]

    # Fit the model using curve_fit
    params, covariance = curve_fit(ggm_model, t, z, p0=initial_guess, maxfev=100_000)

    # Extract the fitted parameters
    K_fit, pc_fit, qc_fit, ps_fit, qs_fit = params

    # Compute the fitted values using the best parameters
    fitted_values = ggm_model(t, *params)

    # Return the fitted parameters and the fitted values
    return K_fit, pc_fit, qc_fit, ps_fit, qs_fit, fitted_values

```

```python
covid_series.reset_index(drop=True, inplace = True)
```

```python
time_series = covid_series
K_fit, pc_fit, qc_fit, ps_fit, qs_fit, fitted_values = fit_ggm_model(time_series)

# Output the fitted parameters
print(f"Fitted parameters: K={K_fit}, pc={pc_fit}, qc={qc_fit}, ps={ps_fit}, qs={qs_fit}")

# Plot the observed vs fitted values
plt.plot(time_series.index, time_series.values, label='Observed')
plt.plot(time_series.index, fitted_values, label='Fitted', linestyle='--')
plt.legend()
plt.title("Fitted GGM vs Observed Data")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

```python
r2 = r2_score(covid_series, fitted_values)  
rmse = np.sqrt(mean_squared_error(covid_series, fitted_values))
print('R2:',r2)
print('RMSE', rmse)
```

```python
# Define the function to predict values using the fitted parameters
def predict_ggm(K, pc, qc, ps, qs, t):
    # Use the fitted model to predict the values for the new time index
    part1 = np.sqrt((1 - np.exp(-(pc + qc) * t)) / (1 + (qc / pc) * np.exp(-(pc + qc) * t)))
    part2 = (1 - np.exp(-(ps + qs) * t)) / (1 + (qs / ps) * np.exp(-(ps + qs) * t))
    return K * part1 * part2

# Create an extended time index (e.g., extending the horizon by 30 more days)
new_time_index = np.arange(time_series.index[-1] + 1, time_series.index[-1] + 365)

# Predict the fitted values for the new time range
extended_fitted_values = predict_ggm(K_fit, pc_fit, qc_fit, ps_fit, qs_fit, new_time_index)

# Optionally, visualize the results
plt.plot(time_series.index, time_series.values, label='Observed (Current Data)')
plt.plot(new_time_index, extended_fitted_values, label='Predicted (Extended Horizon)', linestyle='--')
plt.legend()
plt.title("Observed vs Predicted GGM Values")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```
