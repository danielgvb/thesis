```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_d = pd.read_csv('../Data/silver/covid_data.csv')
df_w = pd.read_csv('../Data/silver/covid_data_weekly.csv')
df_m = pd.read_csv('../Data/silver/covid_data_monthly.csv')
```

# Bass Models

```python
from scipy.optimize import curve_fit
# Define the Bass diffusion model equation
def bass_model(t, p, q, M):
    A = (p + q) / M
    B = p * q / M
    return M * A * (1 - np.exp(-A * t)) / (1 + (B / A) * (np.exp(-A * t)))
```

```python
df_col_m = df_m[df_m['country'] == 'Colombia']
df_col_m.head()
```

```python
time = np.arange(0 , len(df_col_m))
cumulative_adopters = df_col_m['cases'].values
```

```python
# normalize adopters
max_adopters = np.max(cumulative_adopters)
cumulative_adopters = cumulative_adopters/max_adopters *100

# Use curve_fit from SciPy to optimize model parameters
params, covariance = curve_fit(bass_model, time, cumulative_adopters, p0=[0.01, 0.3, 100])

p_optimized, q_optimized, M_optimized = params


# Print the optimized parameters
print('p:', p_optimized)
print('q',q_optimized)
print('M:',M_optimized)

# Predict future adoption with the optimized parameters
future_time = np.arange(0, len(df_col_m))
predicted_adopters = bass_model(
    future_time, p_optimized, q_optimized, M_optimized)
```

```python
plt.plot(predicted_adopters, color='r', label = 'predicted')
plt.scatter(time,cumulative_adopters, label ='fitted')
plt.legend()
plt.title('Bass Model')
plt.show()
```

```python
from sklearn.metrics import r2_score

# Ensure predicted_adopters and cumulative_adopters are NumPy arrays
predicted_adopters = np.array(predicted_adopters)
cumulative_adopters = np.array(cumulative_adopters)

# Compute R² score
r2 = r2_score(cumulative_adopters, predicted_adopters)

# Print R² value
print(f'R² Score: {r2:.4f}')
```

```python
from sklearn.metrics import mean_squared_error

# Un-normalize the predicted adopters
predicted_adopters_unnormalized = (predicted_adopters / 100) * max_adopters

# Un-normalize the cumulative adopters (if needed)
cumulative_adopters_unnormalized = (cumulative_adopters / 100) * max_adopters

# Compute RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(cumulative_adopters_unnormalized, predicted_adopters_unnormalized))

# Output RMSE
rmse
```

# Covid Montly All

```python
r2_bm = {}
rmse_bm = {}

countries = df_m['country'].unique()
num_countries = len(countries)

# Define the grid size (5 columns, 2 rows)
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 8)) 
axes = axes.flatten()  # Flatten to 1D array for easy indexing

for idx, country in enumerate(countries):
    if idx >= 10:  # Avoid exceeding the number of subplots
        break
    
    ax = axes[idx]  # Select subplot
    
    df_temp = df_m[df_m['country'] == country]
    time = np.arange(0, len(df_temp))
    cumulative_adopters = df_temp['cases'].values
    
    # Normalize adopters
    max_adopters = np.max(cumulative_adopters)
    cumulative_adopters = cumulative_adopters / max_adopters * 100
    
    # Use curve_fit from SciPy to optimize model parameters
    params, covariance = curve_fit(bass_model, time, cumulative_adopters, p0=[0.01, 0.3, 100])
    p_optimized, q_optimized, M_optimized = params
    
    # Predict future adoption with the optimized parameters
    future_time = np.arange(0, len(df_temp))
    predicted_adopters = bass_model(future_time, p_optimized, q_optimized, M_optimized)
    
    # Plot in the selected subplot
    ax.plot(predicted_adopters, color='r', label='Predicted')
    ax.scatter(time, cumulative_adopters, label='Fitted')
    ax.legend()
    ax.set_title(f'Bass Model {country}')
    
    # Compute R² score
    r2 = r2_score(cumulative_adopters, predicted_adopters)
    r2_bm[country] = r2
    
    # Un-normalize predicted adopters
    predicted_adopters_unnormalized = (predicted_adopters / 100) * max_adopters
    cumulative_adopters_unnormalized = (cumulative_adopters / 100) * max_adopters
    
    # Compute RMSE
    rmse = np.sqrt(mean_squared_error(cumulative_adopters_unnormalized, predicted_adopters_unnormalized))
    rmse_bm[country] = rmse

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
# Create DataFrame from the dictionaries
df_metrics_bm = pd.DataFrame({
    'country': list(r2_bm.keys()),
    'r2': list(r2_bm.values()),
    'rmse': list(rmse_bm.values())
})
df_metrics_bm
```
