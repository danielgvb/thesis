# Models

```python
import scipy
import matplotlib.pyplot as plt

def bass_model(t, p, q, m):
    """
    Bass diffusion model function.
    
    Parameters:
    - t: Time (days)
    - p: Coefficient of innovation (external influence)
    - q: Coefficient of imitation (internal influence)
    - m: Market potential (total possible cases)
    
    Returns:
    - Cumulative adopters at time t
    """
    return m * (1 - np.exp(-(p + q) * t)) / (1 + (q / p) * np.exp(-(p + q) * t))
```

```python
# Load your transformed data
df = df_long.copy()

# Ensure data is sorted by date
df = df.sort_values('Date')

# Convert dates to days since start
df['Days'] = (df['Date'] - df['Date'].min()).dt.days

# Define time (X) and cases (Y)
X = df['Days'].values  # Time in days
Y = df['Cases'].values  # Cumulative cases

# Normalize Y to avoid numerical instability
m_max = Y.max()
Y = Y / m_max
```

```python
# Initial parameter guesses: p, q, m
initial_guess = [0.01, 0.1, 1]  # Adjust as needed

# Curve fitting to estimate parameters
params, _ = scipy.optimize.curve_fit(bass_model, X, Y, p0=initial_guess, maxfev=10000)

# Extract fitted parameters
p_fitted, q_fitted, m_fitted = params

print(f"Estimated Parameters:\n p = {p_fitted:.4f}, q = {q_fitted:.4f}, m = {m_fitted:.4f}")
```

```python
# Generate future predictions
days_future = np.arange(0, X.max() + 100)  # Predict 100 days ahead
predicted_cases = bass_model(days_future, p_fitted, q_fitted, m_fitted) * m_max  # Scale back

# Plot results
plt.figure(figsize=(10, 5))
plt.scatter(X, Y * m_max, label="Actual Cases", color="blue")
plt.plot(days_future, predicted_cases, label="Bass Model Prediction", color="red")
plt.xlabel("Days Since First Case")
plt.ylabel("Total Cases")
plt.legend()
plt.title("Bass Model Fit for COVID-19 Cases")
plt.show()
```

## GGM

```python
def ggm_model(t, p, q, gamma, m):
    """
    Guseo-Guidolin Model (GGM) for diffusion.
    
    Parameters:
    - t: Time (days)
    - p: Innovation coefficient (external influence)
    - q: Imitation coefficient (internal influence)
    - gamma: Flexibility parameter (modifies carrying capacity over time)
    - m: Maximum adoption potential (total cases)
    
    Returns:
    - Cumulative adopters at time t
    """
    return m * (1 - np.exp(-(p + q) * t)) / ((1 + (q / p) * np.exp(-(p + q) * t)) ** (1 / gamma))
```

```python
# Initial parameter guesses: p, q, gamma, m
initial_guess = [0.01, 0.1, 1.0, 1]  # Adjust as needed

# Curve fitting to estimate parameters
params, _ = scipy.optimize.curve_fit(ggm_model, X, Y, p0=initial_guess, maxfev=10000)

# Extract fitted parameters
p_fitted, q_fitted, gamma_fitted, m_fitted = params

print(f"Estimated Parameters:\n p = {p_fitted:.4f}, q = {q_fitted:.4f}, Î³ = {gamma_fitted:.4f}, m = {m_fitted:.4f}")
```

```python
# Generate future predictions
days_future = np.arange(0, X.max() + 100)  # Predict 100 days ahead
predicted_cases = ggm_model(days_future, p_fitted, q_fitted, gamma_fitted, m_fitted) * m_max  # Scale back

# Plot results
plt.figure(figsize=(10, 5))
plt.scatter(X, Y * m_max, label="Actual Cases", color="blue")
plt.plot(days_future, predicted_cases, label="GGM Prediction", color="red")
plt.xlabel("Days Since First Case")
plt.ylabel("Total Cases")
plt.legend()
plt.title("GGM Fit for COVID-19 Cases")
plt.show()
```
