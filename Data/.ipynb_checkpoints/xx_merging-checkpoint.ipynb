```python
import pandas as pd
import numpy as np
```

# Daily merging

```python
#covid
df_covid = pd.read_csv('silver/covid_data.csv')
df_covid.head()
```

```python
df_covid.country.unique()
```

```python
df_covid['date'] = pd.to_datetime(df_covid['date'])
df_covid['date_str'] = df_covid['date'].astype(str)
```

```python
df_atmos = pd.read_csv('silver/atmosferic_data.csv')
df_atmos.head()
```

```python
df_atmos['date'] = pd.to_datetime(df_atmos['Date'])
df_atmos['date_str'] = df_atmos['date'].astype(str)
```

```python
# rainfall
df_rain = pd.read_csv('silver/rainfall.csv')
df_rain.head()
```

```python
df_rain.Country.unique()
```

```python
df_rain['DATE'] = pd.to_datetime(df_rain['DATE'])
df_rain['date_str'] = df_rain['DATE'].astype(str)
df_rain.rename(columns={'Country':'country'}, inplace = True)
```

```python
# mobility
df_mob = pd.read_csv('silver/mobility_index.csv')
df_mob.head()
```

```python
df_mob['Date'] = pd.to_datetime(df_mob['Date'])
df_mob['date_str'] = df_mob['Date'].astype(str)
```

```python
dict_countries = {'CHL':'Chile', 'COL':'Colombia', 'ITA':'Italy',
                  'MEX':'Mexico', 'PAN':'Panama', 'USA':'US'}
```

```python
dict_countries_2 = {'BR':'Brazil','CL':'Chile', 'CO':'Colombia', 'IT':'Italy',
                  'MX':'Mexico', 'PA':'Panama', 'US':'US', 'DE': 'Germany'}
```

```python
# new countries mobility
new_countries = []
for i in range(len(df_mob)):
    new_country = dict_countries[df_mob['country'][i]]
    new_countries.append(new_country)
```

```python
df_mob['country'] = new_countries
```

```python
# new countries atmospheric
new_countries = []
for i in range(len(df_atmos)):
    new_country = dict_countries_2[df_atmos['Country'][i]]
    new_countries.append(new_country)

df_atmos['country'] = new_countries
```

```python
df_atmos
```

```python
df_atmos_wide = pd.pivot_table(df_atmos, index=['date_str', 'country'], columns = ['Specie'], values = 'median').reset_index()
df_atmos_wide.head()
```

# Merge

```python
df_merge0 = df_covid.merge(df_atmos_wide, how ='left', on=['date_str', 'country'])
df_merge0.head()
```

```python
df_merge1 = df_merge0.merge(df_rain, how ='left', on=['date_str', 'country'])
df_merge1.head()
```

```python
df_merge2 = df_merge1.merge(df_mob,  how ='left', on=['date_str', 'country'])
df_merge2.head()
```

```python
#df_merge1.dropna(inplace = True)
df_merge2.columns
```

```python
df_merge_final = df_merge2[['date', 'country','cases', 'new_cases', 'PRCP', 'TEMP', 'Mean_distance_avg_14day_movavg',
                           'co', 'd', 'dew', 'humidity', 'no2', 'o3', 'pm10', 'pm25',
                            'precipitation', 'pressure', 'so2', 'temperature', 'uvi', 'wd',
                            'wind gust', 'wind speed', 'wind-gust', 'wind-speed']]
```

```python
df_merge_final.columns = ['date', 'country','cases', 'new_cases', 'rain', 'temp', 'dist_14d_avg',
                          'co', 'd', 'dew', 'humidity', 'no2', 'o3', 'pm10', 'pm25',
                          'precipitation', 'pressure', 'so2', 'temperature', 'uvi', 'wd',
                          'wind gust', 'wind speed', 'wind-gust', 'wind-speed']
```

```python
df_merge_final.dropna(inplace = True)
df_merge_final
```

```python
df_merge_final.info()
```

```python
nan_counts_sorted = df_merge_final.isna().sum().sort_values(ascending=False)
print(nan_counts_sorted)
```

```python
df_merge_final_s = df_merge_final[['date', 'country','cases', 'new_cases', 'rain', 'temp', 'dist_14d_avg',
                          'co', 'humidity', 'no2', 'o3', 'pm10', 'pm25','so2']]
```

```python
df_merge_final_s.info()
```

```python
import os

# Specify the directory path
directory_path = 'gold'

# Create the directory if it does not exist
os.makedirs(directory_path, exist_ok=True)

df_merge_final_s.to_csv('gold/covid_daily_multivariate.csv', index = False)
```
