```python
import os
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from tqdm import tqdm
```

```python
# Define file paths
shapefile_path = "WB_countries_Admin0_10m.shp"
csv_directories = ["2020","2021", "2022", "2023"]  # Replace with actual directory path

# Load the world shapefile
world = gpd.read_file(shapefile_path)

# Define selected countries
selected_countries = ['Brazil', 'Chile', 'Colombia',
                      'Italy', 'Panama', 'Mexico',
                      'Uruguay', 'Dominican Republic', 'Germany', 'United States']

# Filter world GeoDataFrame for selected countries
selected_world = world[world['NAME_EN'].isin(selected_countries)]
```

```python
# Create an empty list to store data
data_list = []

# Iterate through multiple directories
for csv_directory_path in csv_directories:
    if not os.path.exists(csv_directory_path):  # Check if directory exists
        print(f"Directory not found: {csv_directory_path}")
        continue

    # Process each CSV file in the directory
    for filename in tqdm(os.listdir(csv_directory_path)):
        if filename.endswith(".csv"):
            file_path = os.path.join(csv_directory_path, filename)

            # ðŸ›‘ Check if the file is empty before reading
            if os.stat(file_path).st_size == 0:
                print(f"Skipping empty file: {filename}")
                continue  # Skip this file

            try:
                # Read CSV, assuming first two rows are metadata and skipping them
                df = pd.read_csv(file_path)

                # ðŸ›‘ Ensure required columns exist before processing
                required_columns = {'LATITUDE', 'LONGITUDE', 'DATE', 'PRCP', 'TEMP'}
                if not required_columns.issubset(df.columns):
                    print(f"Skipping file {filename}: Missing required columns")
                    continue  # Skip this file

                # Extract relevant columns
                df = df[['DATE', 'LATITUDE', 'LONGITUDE', 'PRCP', 'TEMP']]

                # Convert coordinates into GeoDataFrame
                df['geometry'] = df.apply(lambda row: Point(row['LONGITUDE'], row['LATITUDE']), axis=1)
                df_gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

                # Spatial join to assign each point to a country
                df_gdf = gpd.sjoin(df_gdf, selected_world[['NAME_EN', 'geometry']], how="inner", predicate="within")

                # Rename country column
                df_gdf = df_gdf.rename(columns={'NAME_EN': 'Country'})

                # Append to the list
                data_list.append(df_gdf)

            except Exception as e:
                print(f"Error processing file {filename} in {csv_directory_path}: {e}")
                continue  # Skip and move to the next file

# Combine all the processed data into a single DataFrame
if data_list:
    final_df = pd.concat(data_list, ignore_index=True)

    # Drop unnecessary columns
    final_df = final_df[['DATE', 'LATITUDE', 'LONGITUDE', 'PRCP', 'TEMP', 'Country']]
```

```python
df_col = final_df[final_df['Country'] == 'Colombia']
df_col_prcp = df_col.groupby(['DATE'])[['PRCP']].sum().reset_index()
df_col_temp = df_col.groupby(['DATE'])[['TEMP']].mean().reset_index()
df_col_merge = df_col_prcp.merge(df_col_temp, on = 'DATE')
df_col_merge['DATE'] = pd.to_datetime(df_col_merge['DATE'])
```

```python
import matplotlib.pyplot as plt
plt.plot(df_col_merge.DATE, df_col_merge.PRCP)
```

```python
plt.plot(df_col_merge.DATE, df_col_merge.TEMP)
```

```python
df_col_merge['TEMP_c'] = (df_col_merge.TEMP - 32) * 5/9
plt.plot(df_col_merge.DATE, df_col_merge.TEMP_c)
```
