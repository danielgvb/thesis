```python
# Directory containing the CSV files
directory_path = '2021'  # Replace with your directory path

# List to store the coordinate pairs
coordinates = []

# Loop through each file in the directory
for filename in tqdm(os.listdir(directory_path)):
    if filename.endswith(".csv"):
        # Construct the full file path
        file_path = os.path.join(directory_path, filename)

        # Read the CSV file, skipping the first two lines and using the third line as the header
        df = pd.read_csv(file_path)

        # Extract the first row of the DataFrame
        first_row = df.iloc[0]

        # Get the LATITUDE and LONGITUDE values
        latitude = first_row['LATITUDE']
        longitude = first_row['LONGITUDE']

        # Check for NaN values and append the coordinate pair to the list if valid
        if pd.isna(latitude) or pd.isna(longitude):
            continue
        else:
            coordinates.append((latitude, longitude))

# Create an object with all the coordinate pairs
coordinates_object = {
    "coordinates": coordinates
}
```

```python
import geopandas as gpd

# Set the environment variable
os.environ["SHAPE_RESTORE_SHX"] = "YES"

# Load the shapefile
world = gpd.read_file("WB_countries_Admin0_10m.shp")

print(world.head())
```

```python
# Extract only country names and their geometry
country_coordinates = world[['NAME_EN', 'geometry']]

# Convert geometries to coordinate lists
country_coordinates['coordinates'] = country_coordinates['geometry'].apply(lambda x: list(x.exterior.coords) if x.geom_type == 'Polygon' else [list(p.exterior.coords) for p in x.geoms])


df_coordinates = pd.DataFrame(country_coordinates[['NAME_EN', 'coordinates']])
df_coordinates.head()
```
