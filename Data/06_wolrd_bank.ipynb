```python
df = pd.read_excel('world_bank/WDIEXCEL.xlsx', sheet_name='Data')
df.head()
```

```python
df_series = pd.read_excel('world_bank/WDIEXCEL.xlsx', sheet_name='series_to_include')
df_countries = pd.read_excel('world_bank/WDIEXCEL.xlsx', sheet_name='countries_to_include')
```

```python
df_countries
```

```python
series_keep = df_series['Series']
countries_dict = df_countries.set_index('Country Code')['Short Name'].to_dict()
countries_dict
```

```python
df = df[df['Indicator Code'].isin(series_keep)]
df = df[df['Country Code'].isin(countries_dict.keys())]
df.head()
```

```python
df = df[['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2019','2020', '2021', '2022', '2023']]
df.head()
```

```python
# coun nan values for 2020 for every country
df['2020'].isna().groupby(df['Country Name']).sum()
```

```python

# Fill NaNs in columns 2021, 2022, and 2023 with the average of the previous periods
for year in ['2021', '2022', '2023']:
    df[year] = df[year].fillna(df.loc[:, '2019':'2020'].mean(axis=1))
```

```python
# drop nans
df = df.dropna(subset=['2019'])
```

```python
df.head()
```

```python
df.info()
```

```python
# wide to long
df_long = pd.melt(df, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='Value')
df_long.head()
```

```python
df_long.columns = ['country', 'country_code', 'indicator_name', 'indicator_code', 'year', 'value']
df_long['indicator_name'].unique()
```

```python
df_long.to_csv('silver/world_bank_data.csv', index=False)
```
